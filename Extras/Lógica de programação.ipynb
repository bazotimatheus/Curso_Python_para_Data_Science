# Criando minha primeira função

```python
def saudacao():
  nome = input('Qual o seu nome? ')
  print(f'Olá, {nome}.')
```

```python
saudacao()
```

```python
def nome_completo():
  nome = input('Qual o seu primeiro nome? ')
  sobrenome = input('Qual o seu sobrenome? ')
  nome_inteiro = nome + ' ' + sobrenome
  print(f'Olá, {nome_inteiro}')

nome_completo()
```

# Condicional

```python
idade = 15

def verifica_se_pode_dirigir(idade):
  if idade >= 18:
    print('Tem permissão para dirigir')
  else:
    print('Não tem permissão para dirigir')

verifica_se_pode_dirigir(idade)
```

# Convertendo tipo para inteiro

```python
def verifica_se_pode_dirigir_sem_parametros():
  idade = input('Qual sua idade? ')
  idade = int(idade)
  if idade >= 18:
    print('Tem permissão para dirigir')
  else:
    print('Não tem permissão para dirigir')

verifica_se_pode_dirigir_sem_parametros()
```

# Indo até a padaria

```python
espaco = input('Qual a distância percorrida? ')
tempo = input('Quanto tempo levou? ')

espaco = float(espaco)
tempo = float(tempo)

def calcula_velocidade(espaco, tempo):
  velocidade = espaco / tempo
  print(f'Velocidade: {velocidade} m/s')

calcula_velocidade(espaco, tempo)
```

```python
def verifica_se_pode_dirigir(idades):
  for idade in idades:
    if idade >= 18:
      print(f'{idade} anos de idade, TEM permissão para dirigir')
    else:
      print(f'{idade} anos de idade, NÃO TEM permissão para dirigir')

verifica_se_pode_dirigir(idades)
```

```python
permissoes = []
idades = [20, 14, 40]

def verifica_se_pode_dirigir(idades, permissoes):
  for idade in idades:
    if idade >= 18:
      permissoes.append(True)
    else:
      permissoes.append(False)

verifica_se_pode_dirigir(idades, permissoes)
```

```python
permissoes
```

```python
for permissao in permissoes:
  if permissao == True:
    print('Tem permissão para dirigir')
  else:
    print('Não tem permissão para dirigir')
```
